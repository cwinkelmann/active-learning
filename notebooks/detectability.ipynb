```python
from active_learning.feature_extraction.point_detectability import SamplingConfig, RGBContrastSampler
from pathlib import Path
```

```python
def get_config(base_path: Path):
    return SamplingConfig(
        csv_path=base_path / 'herdnet_format_512_0_crops.csv',
        image_dir=base_path / 'crops_512_numNone_overlap0',
        output_dir=base_path / 'outputs',
        buffer_radius=5,
        patch_size=32,
        min_distance=50,
        object_size_pixels=8,
        max_images=1500,
        visualize_first_n=2,
    )
```

```python


"""Main entry point."""
base_path = Path('/raid/cwinkelmann/training_data/iguana/2025_10_11/Floreana_detection/train')



config = get_config(base_path)

sampler = RGBContrastSampler(config)
summary_df = sampler.run()

print(summary_df)
```

```python
from IPython.display import Image, display

# plot the summary plot
plot_path = base_path / "outputs/dataset_summary_plot.png"


# Display the image
display(Image(filename=plot_path))
```
